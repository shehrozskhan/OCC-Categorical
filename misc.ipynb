```python
import openml

# Fetch dataset list
datasets = openml.datasets.list_datasets(output_format="dataframe")

# Filter datasets with only categorical features and exactly 2 classes
categorical_binary_datasets = datasets[
    (datasets['NumberOfNumericFeatures'] == 0) &  # No numeric features
    (datasets['NumberOfClasses'] == 2) & # Only two classes
    (datasets['NumberOfInstances'] > 100) &  # More than 100 instances
    (datasets['NumberOfInstances'] < 5000) & # Less than 5000 instances
    (datasets['NumberOfFeatures'] >= 5)  # At least 5 features
]

# Display dataset ID and name
print("datasets=", categorical_binary_datasets.shape[0])
print(categorical_binary_datasets[['did', 'name']].to_string())
```
